11. What is the Mean value of each column against each 'Weather Condition ? 

```python
import pandas as pd

data = pd.read_csv("weather_detailed_dataset.csv")
numeric_columns = data.select_dtypes(include=['number'])
mean_values = data.groupby("Weather")[numeric_columns.columns].mean()

print(mean_values)
```
